### Resolving the issue with displaying the crime time
In my project, I encountered a problem with correctly displaying the event times, as they were stored in two columns: "Date OCC" and "Time OCC". Additionally, in the "Time OCC" column, the time was stored in an incorrect format, which caused errors when displaying data in Power BI.
To solve this problem, I created a new column named "DATE TIME OCC" which combined the date and time of the event into one column. This allowed me to obtain the correct date and time format, which greatly facilitated further work with the data.


```python
df[df['TIME OCC']>2300].head(5)
```

```python
df["DATE TIME OCC"] = df["TIME OCC"].apply(str)
```

```python
df["DATE TIME OCC"] = df["DATE TIME OCC"].apply(lambda x : x.zfill(4))
```

```python
df["DATE TIME OCC"] = df["DATE TIME OCC"].apply(lambda A : A[:2] + ':' + A[2:])
```

```python
df["TIME OCC"] = df["DATE TIME OCC"]
```

```python
df["DATE TIME OCC"] = df["DATE OCC"].apply(lambda x : x[:10]) +" "+ df["DATE TIME OCC"]
```

```python
df["Date Rptd"] = df["Date Rptd"].apply(lambda x : x[:10])
```

### Conversion of  Codes to Ethnic Group Names
I cleaned up the "Vict Descent" column containing data on the ethnicity of victims by converting coded letters into more easily understandable ethnic names.

```python
print(df["Vict Descent"].unique())
```

```python
ethnic_groups = {"A": "Asian", "B": "Black", "C": "Other", "D": "Asian", "F": "Asian", "G": "Other", "H": "Latino", "I": "Other", "J": "Asian", "K": "Asian", "L": "Asian", "O": "Other", "P": "Other", "S": "Other", "U": "Other", "V": "Asian", "W": "White", "Z": "Asian"}

df["Vict Descent"] = df["Vict Descent"].replace(ethnic_groups)
```
